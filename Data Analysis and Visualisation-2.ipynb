```python
import pandas
df=pandas.read_csv('reviews.csv')
df
```

```python
df['Timestamp']=pandas.to_datetime(df['Timestamp'])
```

```python
df1=df.set_index('Timestamp',drop=False)
```

```python
#Creating column with date attribute 
df['Day']=df['Timestamp'].dt.date
```

```python
df1['Day']=df['Timestamp'].dt.date
df1
```

```python
df_grp=df.groupby(['Day']).mean()
df_grp
```

## 3. Turning data into visualisation

#### Q1. What is the daily average ratings of courses ?

```python
#Daily Rating of the course
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,5))
plt.plot(df_grp.index,df_grp.Rating)
```

#### Q2. What are average weekly rating of courses ?

```python
# creating coloumn week with converting the timestamp to week 
df['Week']=df['Timestamp'].dt.strftime('%Y %U')
df
```

```python
df_grp2=df.groupby('Week').mean()
df_grp2
```

```python
# Weekly ratings of course
plt.figure(figsize=(30,10))
plt.plot(df_grp2.index,df_grp2.Rating)
```

#### Q3. What are average monthly rating of courses ?

```python
# creating coloumn month with converting the timestamp to month 
df['Month']=df['Timestamp'].dt.strftime('%Y %m')
df
```

```python
month_avg=df.groupby('Month').mean()
month_avg
```

```python
# Monthly ratings of course
plt.figure(figsize=(30,7.5))
plt.plot(month_avg.index,month_avg.Rating)
plt.show()
```

#### Q4. What are average monthly ratings of each course ?

```python
df
```

```python
monthcrs_avg=df.groupby(['Month','Course Name']).mean().unstack()
monthcrs_avg
```

```python
# Monthly ratings of every course
plt.figure(figsize=(30,5))
for i in monthcrs_avg['Rating'].columns:
    plt.plot(monthcrs_avg.index,monthcrs_avg['Rating'][i],label=i)
plt.legend()
```

#### Q5. What is happiest day of week ?


```python
df
```

```python
# creating coloumn weekday with converting the timestamp to days 
df['Weekday']=df['Timestamp'].dt.strftime('%A')
df
```

```python
weekday_avg=df.groupby('Weekday').mean()
weekday_avg
```

```python
days=['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
weekday_avg=weekday_avg.reindex(days)
weekday_avg
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('Days')
plt.ylabel('Rating')
plt.plot(weekday_avg.index,weekday_avg.Rating)
# It is observed that Friday is the happiest day of the week.
```
